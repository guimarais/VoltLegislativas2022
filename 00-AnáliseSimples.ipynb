```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
```

### Candidatos por Circulo

```python
fig, ax = plt.subplots(figsize=(16,6))
df[['Lugar', 'Circulo']].groupby(by='Circulo').count().sort_values('Lugar', ascending=False).plot.bar(ax=ax)
ax.set_ylabel("Candidatos")
plt.show()
```
